```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
```

Omni anomaly performance in the paper



# SMD

```python
# default params
folder = 'result_SMD_1.2_GD'

with open(f'./{folder}/result.json', 'r') as file:
    result = json.load(file)

valid_loss = pd.read_csv(f'./{folder}/training_valid_loss.csv')
loss = pd.read_csv(f'./{folder}/training_loss.csv')
steps = np.arange(100, (len(loss)+1)*100, 100)

plt.plot(steps, valid_loss, label="valid")
plt.plot(steps, loss, label='train')
plt.legend()
plt.title(f'Precision: {result["precision"]:.4f}; Recall: {result["recall"]:.4f}')
plt.xlabel('step')
plt.ylabel('loss');
```

# Training and validation loss comparison

```python
adam = pd.read_csv('./result_SMD_1.2_default/training_loss.csv')
adadelta = pd.read_csv('./result_SMD_1.2_Adadelta/training_loss.csv')
adagrad = pd.read_csv('./result_SMD_1.2_Adagrad/training_loss.csv')
gd = pd.read_csv('./result_SMD_1.2_GD/training_loss.csv')
rmsprop = pd.read_csv('./result_SMD_1.2_RMSProp/training_loss.csv')
nesterov = pd.read_csv('./result_SMD_1.2_Nesterov/training_loss.csv')
momentum = pd.read_csv('./result_SMD_1.2_Momentum/training_loss.csv')

steps = np.arange(100, 3300, 100)

plt.plot(steps, adam, label="Adam")
plt.plot(steps, adadelta, label="Adadelta")
plt.plot(steps, adagrad, label="Adagrad")
plt.plot(steps, gd, label="Gradient Descent")
plt.plot(steps, rmsprop, label="RMS Prop")
plt.plot(steps, nesterov, label="Nesterov Momentum (0.5)")
plt.plot(steps, momentum, label="Momentum (0.5)")
plt.legend()
plt.title(f'Optimizers training loss')
plt.xlabel('step')
plt.ylabel('loss');
```
